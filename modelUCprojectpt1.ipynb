```python
# index is the index -> i.e the datestop,coordinates etc, m is the number of neighbours to consider, we'll go with 4
# dont separate x and y yet
def get_neighbours(df):
    for index in range(len(df)):
        #df.coordinates=df.coordinates.apply(ast.literal_eval) #gives error if its already changed to it?!
        coord=df.iloc[index]['coordinates']
        date=df.iloc[index]['datestop']
        #print('examined entry is ',coord,date)
        #first filter on the same date all the values
        df_dates=df.loc[df['datestop']==date]
        #create the m neighbouring coordinate tuples - and add their indexes from df_dates indexes if they exist
        indexes=[]
        have_neigh=[False,False,False,False]
        #N1 is x,y+1, N2=x-1,y, N3=x+1,y, N4=x,y-1
        if(coord[1]<24):
            N1=(coord[0],coord[1]+1)
            idx=df_dates.index[df_dates['coordinates']==N1].tolist()
            #print('fetched values are for N1',f'{N1} ',idx)
            if len(idx) > 0:
                indexes.append((idx[0],'N1'))
                have_neigh[2]=True
            #print('length indexes for N1',f'{N1} is',len(idx))

        if(coord[0]>=1):
            N2=(coord[0]-1,coord[1]) 
            idx=df_dates.index[df_dates['coordinates']==N2].tolist()
            #print('fetched values are for ',f'{N2} ',idx)
            if len(idx) > 0:
                indexes.append((idx[0],'N2'))
                have_neigh[0]=True
            #print('and length indexes for N2',f'{N2} is',len(idx)) 

        if (coord[0]<23): #double check this one if 
            N3=(coord[0]+1,coord[1])
            idx=df_dates.index[df_dates['coordinates']==N3].tolist()
            #print('fetched values are for N3',f'{N3} ',idx)
            if len(idx) > 0:
                indexes.append((idx[0],'N3'))
                have_neigh[3]=True
            #print('and length indexes N3',f'{N3} is',len(idx))
        
        if(coord[1]>=1):
            N4=(coord[0],coord[1]-1)
            idx=df_dates.index[df_dates['coordinates']==N4].tolist()
            #print('fetched values are for ',f'{N4} ',idx)
            if len(idx) > 0:
                indexes.append((idx[0],'N4'))
                have_neigh[1]=True
            #print('and length indexes for N4',f'{N4} is',len(idx)) 
        
        #now let's get the values for our neighbours X and Y - into separate columns for each neighbour
        #print('and the neighbours')
        #print(have_neigh)
        for idx in range(len(indexes)):
            #print('neighbour is: ',indexes[idx]) 
            vals=df_dates.loc[indexes[idx][0]][['SAF_COUNT','311_complaints_COUNT','nypd_compaints_COUNT']].values
            #print('values for entry with index ',f'{index} is: ',vals,'with type',type(vals))
            #print('Neighbour is: ',indexes[idx][1])
            #print(f'{indexes[idx][1]}_311 that should have the value ',vals[1])
            df.at[index,f'{indexes[idx][1]}_SAF']=vals[0] 
            df.at[index,f'{indexes[idx][1]}_311']=vals[1]
            df.at[index,f'{indexes[idx][1]}_nypd']=vals[2]
    return df
```
