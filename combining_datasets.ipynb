```python
import pandas as pd
import numpy as np
import ast
import os
```

```python
if os.name == 'posix':
    grouped_location = '/Users/ernestvmo/OneDrive - Universiteit Leiden/Q1/Urban Computing/Project/data/grouped_csv/'
else:
    grouped_location = 'C:/Users/Ernest/OneDrive - Universiteit Leiden/Q1/Urban Computing/Project/data/grouped_csv/'

saf_df = pd.read_csv(grouped_location + 'saf_grouped.csv', index_col=[0])
nypd_df = pd.read_csv(grouped_location + 'grouped_nypd_complaints.csv', index_col=[0])
_311_df = pd.read_csv(grouped_location + 'grouped_311.csv', index_col=[0])
# taxi_df = pd.read_csv(grouped_location + 'grouped_taxi.csv', index_col=[0])
taxi_pickup = pd.read_csv(grouped_location + 'taxi_pickup.csv', index_col=0)
taxi_dropoff = pd.read_csv(grouped_location + 'taxi_dropoff.csv', index_col=0)

nypd_df = nypd_df.rename(columns={'nypd_compaints_COUNT':'nypd_complaints_COUNT'})

del taxi_pickup['Unnamed: 0.1'], taxi_dropoff['Unnamed: 0.1']
```

### SAF to NYPD

```python
saf_merge = nypd_df.merge(saf_df, how='left', on=['datestop','coordinates'])
```

```python
saf_merge.fillna(0, inplace=True)
```

```python
saf_merge
```

### 311 to NYPD

```python
_311_merged = nypd_df.merge(_311_df, how='left', on=['datestop', 'coordinates'])
```

```python
_311_merged = _311_merged.replace(np.nan, 0)
```

```python
_311_merged
```

### Taxi

##### Pickup

```python
taxi_pickup = taxi_pickup.rename(columns={'pickup_datetime':'datestop'})
```

```python
taxi_pickup.sort_values(by=['datestop', 'coordinates'], inplace=True)
```

```python
taxi_pickup_merged = nypd_df.merge(taxi_pickup, how='left', on=['datestop','coordinates'])
```

```python
taxi_pickup_merged = taxi_pickup_merged.fillna(0)
```

```python
taxi_pickup_merged
```

##### Dropoff

```python
taxi_dropoff = taxi_dropoff.rename(columns={'dropoff_datetime':'datestop'})
```

```python
taxi_dropoff.sort_values(by=['datestop', 'coordinates'], inplace=True)
```

```python
taxi_dropoff_merged = nypd_df.merge(taxi_dropoff, how='left', on=['datestop','coordinates'])
```

```python
taxi_dropoff_merged = taxi_dropoff_merged.fillna(0)
```

```python
taxi_dropoff_merged
```

### Combine ALL

```python
taxi_merged = taxi_pickup_merged.merge(taxi_dropoff_merged, on=['datestop', 'coordinates', 'nypd_complaints_COUNT'])
```

```python
taxi_merged = taxi_merged.rename(columns={'passenger_count_x':'passenger_count_pickup', 'passenger_count_y':'passenger_count_dropoff', 'total_amount_x':'total_amount_pickup','total_amount_y':'total_amount_dropoff'})
```

```python
saf_311_merged = saf_merge.merge(_311_merged, how='inner', on=['datestop','coordinates', 'nypd_complaints_COUNT'])
```

```python
all_merged = saf_311_merged.merge(taxi_merged, how='inner', on=['datestop','coordinates', 'nypd_complaints_COUNT'])
```

```python
all_merged = all_merged[['datestop','coordinates','SAF_COUNT','311_complaints_COUNT','passenger_count_pickup','total_amount_pickup','passenger_count_dropoff','total_amount_dropoff','nypd_complaints_COUNT']]
```

```python
all_merged
```

```python
all_merged.coordinates = all_merged.coordinates.apply(ast.literal_eval)
```

```python
if os.name == 'posix':
    save_location = '/Users/ernestvmo/OneDrive - Universiteit Leiden/Q1/Urban Computing/Project/data/FINAL DATA/'
else:
    save_location = 'C:/Users/Ernest/OneDrive - Universiteit Leiden/Q1/Urban Computing/Project/data/FINAL DATA/'

all_merged.to_csv(save_location + 'dataset.csv')
```
