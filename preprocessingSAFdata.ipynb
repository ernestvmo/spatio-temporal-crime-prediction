## Now need to filter on datestop to only cover our interessting dates: july 1 2012 to june 30 2013.
format is date, month,year so values date >=1, month >=7 for 2012 and month<=6 for 2013. This will be done by converting into the generic date-format.

```python
def format_date(dataframe, column):
    new_date = []
    format_date = '%d/%m/%Y'
    for fulldate in dataframe[column].values.astype('str'):
        year, fulldate = int(fulldate[-4:]), fulldate[:-4] 
        day, fulldate = int(fulldate[-2:]), fulldate[:-2] 
        month = int(fulldate[:])
        new_date.append(f'{day}/{month}/{year}')
    dataframe[column] = pd.to_datetime(new_date, format=format_date)
    return dataframe
```

```python
df_2012_dates=format_date(df_2012,column='datestop')
df_2012_dates.head()
```

```python
df_2013_dates=format_date(df_2013,column='datestop')
df_2013_dates.head()
```

```python
#apply format_date method on the column datestop dont change anything else
d_2012=df_2012_dates.loc[df_2012_dates['datestop']>='2012-07-01']
d_2012.head()
```

```python
d_2013=df_2013_dates.loc[df_2013_dates['datestop']<='2013-06-30']
d_2013.head()
```

```python
saf=pd.concat([d_2012,d_2013])
saf.head()
```

```python
#make it into a file
saf.to_csv('NYPDstopAndFrisk12-13.csv')
```
